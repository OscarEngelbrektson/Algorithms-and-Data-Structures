```python
def lcs(x,y):
    m = len(x)
    n = len(y)
    
    c = [[0 for col in range(n+1)] for row in range(m+1)]
    b = [[0 for col in range(1,n+1)] for row in range(1,m+1)]
    
    for i in range(1, m+1):
        
        for j in range(1, n+1):
            
            #Case 1
            if x[i-1] is y[j-1]:
                c[i][j] = c[i-1][j-1]+1
                b[i-1][j-1] = "diagonal"
            
            #Case 2
            elif c[i-1][j] >= c[i][j-1]:
                c[i][j] = c[i-1][j]
                b[i-1][j-1] = "up"
            
            #Case 3
            else:
                c[i][j] = c[i][j-1]
                b[i-1][j-1] = "left"
    return c, b

def print_lcs(b, x, i, j):
    if i is 0 or j is 0:
        return
    if b[i][j] is "diagonal":
        print_lcs(b, x, i-1, j-1)
        print(x[i])
    elif b[i][j] is "up":
        print_lcs(b, x, i-1, j)
        print(x[i])
    else:
        print_lcs(b, x, i, j-1)
```

```python
def lis(x): 
    n = len(x) 
  
    longest = [1 for _ in range(x)]
    
    for i in range (1 , n): 
        for j in range(0 , i): 
            if x[i] > x[j] and x[i]< x[j] + 1 : 
                longest[i] = longest[j]+1
  
    maximum = 0
  
    for i in range(n): 
        maximum = max(maximum , longest[i]) 
  
    return maximum
```
