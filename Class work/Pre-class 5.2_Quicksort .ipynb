```python
# 7.2-2

'''
In my implementation, for every j, if a[j] <= pivot we swap elements and increment the pivot marker.
If all elements are = to eachother, then the pivot will always be at the end of the array.
Creating subarrays of length 0 and n-1 for each "divide" step.
This means the recursion tree will have depth n.
Thus the complexity is O(n^2), the worst case.
'''

# 7.2-3

'''
In my implementation, if a[j] <= pivot I increment and swap.
Thus, if it is sorted in decreasing order, then the pivot will be the smallest number as it is the last element in the array.
So the "smaller" index will never be incremented and yet again we will have
sublists of 0 and n-1. -> Same asymptotic behavior as previous question.
'''
```

```python
def partition(a, low, high):
    smaller = low-1 #Index of smaller element
    pivot = a[high] #Pivot is last element in subarray
    
    for j in range(low, high):
        
        #Is element smaller than pivot
        if a[j] <= pivot:
    
            smaller += 1  #Increment smaller
            a[smaller], a[j] = a[j], a[smaller] #Swap
            
    # Put pivot in correct position
    a[high], a[smaller+1] = a[smaller+1], a[high] 
        
    #Return position of pivot
    return smaller+1

def quick_sort(a, low, high):
    
    if low < high:
        #Find partition
        piv = partition(a, low, high)
        
        #Call quicksort recursively on each side of pivot
        quicksort(a, low, piv-1)
        quicksort(a, piv+1, high)
    
    #Return sorted array
    return a
```

```python
#Check if quicksort works by comparing to built-in sort function
import numpy as np

def quick_sort_checker():
    j = 0
    correct = True
    
    #Check up to 50 times
    while j <= 50 and correct:
        
        #On random list of length between 0 and 1000, 
        #with values between -100 and 100
        test_array = list(np.random.randint(
            -100, 100, np.random.randint(0, 1000, 1)))
        
        #Sort using built-in function
        a = sorted(test_array)
        
        #If sorted arrays don't match, return False
        if a != quick_sort(test_array, 0, len(test_array)-1):
            correct = False
        
        #increment j
        j += 1
        
    return correct

quick_sort_checker()
```

```python
#Fibonacci vs. Quicksort

def fib(n):
    if n <= 0:
        return 0
    if n == 1:
        return 1
    return fib(n-1) + fib(n-2)

start = time.time()
fib(100)
end = time.time()

time_fib100 = end-start
print(time_fib100)

'''
This is taking a while so I will analyse it theoretically.
The nth number in the sequence is the sum of the two pervious numbers
This gives the recurrence relation,
F(N) = F(N-1) + F(N-2) + O(1)
where O(1) represents the constant time addition.

As we call the function twice for every n, the recursion tree
will have 2^n leaf nodes. This gives the algorithm time complexity
O(2^n).
2^100 is an enormous number, 
I understand why my code is taking a long time to execute.

As T(100^2) < T(2^100), I prefer running Quicksort.
'''
```
